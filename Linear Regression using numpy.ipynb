# Linear Regression using numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area  = [1.2,1.8,2.4,3.5,6.4,5.2,8.4,7.6,8.5,9.2,5.6,3.8,4.8,2.7,3.1]
price = [210,230,220,340,590,540,780,820,810,860,520,420,520,320,250]
len(area),len(price)
```

```python
plt.scatter(area,price)
plt.show()
```

```python
x = np.array(area)
y = np.array(price)
```

```python
errors = []
m = 0.5
c = 0.1
LR = 0.002
for i in range(500):
    # get the predictions
    yhat = m*x + c
    
    # calculate error
    e = np.mean(np.square(y-yhat))
    
    # apply gradient descent
    dedm = 2*np.dot((yhat-y),x)
    dedc = 2*np.dot((yhat-y),np.ones(len(y)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(e)
    errors.append(e)
    plt.figure(figsize=(8,6))
    plt.scatter(x,y,c='b')
    plt.plot(x,yhat,c='r')
    plt.show()
```

```python
print(m,c)
```

```python
plt.figure(figsize=(20,8))
plt.scatter(np.arange(0,500),errors)
plt.show()
```
