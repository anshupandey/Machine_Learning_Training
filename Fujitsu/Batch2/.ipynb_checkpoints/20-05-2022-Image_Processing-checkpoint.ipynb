# Image Processing

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Images
    - images are composed of pixels, images can be thought of as collection of pixels
    - each pixel represents a color, one pixel = one color box
    - primarily images are processed in 8 bit size. that means values in pixels can vary from 0000 0000 (0) to 1111 1111 (255)
    - images can be of two types
        - GrayScale
        - Color (RGB)
        
        
### GrayScale images
    - 1D images
    - each pixel can be represented with one value - proportion of white color
    - each pixel value varies from 0 to 255, 0 = black color, 255 = white color

```python
x = np.array([[0,255,0,255],
              [0,255,0,255],
              [0,255,0,255],
              [0,255,0,255]])
x.shape
```

```python
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,255,0,255],
              [255,0,255,0],
              [0,255,0,255],
              [255,0,255,0]])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[255,0,0,255],
              [255,0,255,0],
              [255,255,0,0],
              [255,0,255,0],
              [255,0,0,255],])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[255,255,255,255],
              [255,0,0,255],
              [255,255,255,255],
              [255,0,0,255],
              [255,0,0,255],])
plt.imshow(x,cmap='gray')
plt.show()
```

## Color images
    - each pixel is represented by 3 values (red,green,blue) each value varying from 0 to 255
    - 3D images (3D matrices), 3 dimensions (rows, cols, channels)
    - [0,0,0] = black color
    - [255,0,0] = Red color
    - [0,0,255] = Blue Color
    - [0,255,0] = Green Color
    - [255,255,0] = yellow color
    - [255,255,255] = white color

```python
y = np.array([[[0,0,0],[0,0,255]],
               [[0,255,0],[250,150,10]],
               [[255,0,0],[255,255,255]],
               [[255,255,0],[0,255,255]],
              ])

y.shape
```

```python
plt.imshow(y)
plt.show()
```

## Working with natural images

    - In case of images, 
    - in geometry dimensions are written as (x,y) or width X height
    - in algebra (matrix/arrays) dimensions are reffered as rows,columns
    
    
    - number of rows = height of the image, in geometry it represents y axis
    - numer of columns = width of the image, in geomtery it represents x axis
    
    for an image having width x height = 525 x 350
    when loaded in programming, shape of array/matrix would be rows x cols = 350 x 525

## Working with openCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"E:\image_data\dhoni-kohli.jpg")
img.shape
```

```python
cv2.imshow("myImage",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
plt.show()
```

```python
# BGR to RGB
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img2.shape
```

```python
plt.imshow(img2)
plt.show()
```

```python
# BGR to GRAY
img3 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img3.shape
```

```python
plt.imshow(img3,cmap='gray')
plt.show()
```
