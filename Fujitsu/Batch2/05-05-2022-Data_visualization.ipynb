## Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temperature = [22,24,25,26,24,28,27,26,29,32,35,34,32,38,39,42,44,45,48,46,38,34,32,28,27,26,22]
len(temperature)
```

```python
humidity = [18,16,14,15,16,15,14,17,12,10,10,8,9,8,7,5,7,6,8,9,10,12,15,16,14,18,16]
len(humidity)
```

```python
plt.plot(temperature)
plt.show()
```

```python
plt.plot(temperature)
plt.plot(humidity)
plt.show()
```

```python
plt.plot(temperature)
plt.show()
plt.plot(humidity)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(temperature)
plt.plot(humidity)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(temperature,c='lightblue',label='temperature')
plt.plot(humidity,c='orange',label='humidity')
plt.title("Temperature and HUmidity")
plt.xlabel("day number")
plt.ylabel("temperature and humidity")
plt.xlim([0,30])
plt.legend()
plt.show()
```

```python
plt.plot(temperature,humidity)
plt.show()
```

```python
# scatterplot
days = np.arange(len(temperature))
```

```python
plt.scatter(days,temperature)
plt.scatter(days,humidity)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(temperature,c='lightblue',label='temperature')
plt.plot(humidity,c='orange',label='humidity')
plt.scatter(days,temperature,c='blue')
plt.scatter(days,humidity,c='red')
plt.title("Temperature and HUmidity")
plt.xlabel("day number")
plt.ylabel("temperature and humidity")
plt.xlim([0,30])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.hist(temperature)
plt.show()
```

```python
# barplot
city = ['DXB','AUH','KUL','MAA','BLR']
vals = [45,56,75,45,52]

plt.figure(figsize=(12,4))
plt.bar(city,vals,color=['lightblue','orange','yellow','lightgreen','red'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(vals,labels=city)
plt.show()
```
