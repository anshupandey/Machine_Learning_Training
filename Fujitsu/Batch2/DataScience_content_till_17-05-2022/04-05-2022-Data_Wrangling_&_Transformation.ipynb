```python
import pandas as pd
import numpy as np
```

```python
# load regiment.csv
df1 = pd.read_csv(r"D:\AI\data\datasets-1\regiment.csv",index_col='index')
df1.shape
```

```python
df2 = pd.read_csv(r"D:\AI\data\datasets-1\regiment2.csv")
df2.shape
```

```python
df1
```

```python
df2
```

```python
# merger two dataframes
df = pd.merge(left=df1,right=df2,on='name',how='inner')
df
```

```python
# create a new column representing improvement made by soldiers during the training
df['improvement'] = df.postTestScore - df.preTestScore
df
```

```python
# converting a numeric to categoric
df['impr_c'] = pd.cut(df.improvement,bins=[0,30,50,90],labels=['low','medium','high'])
df
```

```python
# using q cut
df['impr_q'] = pd.qcut(df.improvement,q=4,labels=['very low','low','high','veryhigh'])
df
```

```python
df['pct'] = df.improvement.rank(pct=True)
df
```

```python
df
```

```python
# dropping columns
df.drop(columns=['impr_c','impr_q','pct'],inplace=True)
df
```

## Apply method

```python
def convert(name):
    return name.upper()
```

```python
convert("Anshu")
```

```python
df.name = df.name.apply(convert)
df
```

```python
df.name = df.name.str.lower()
df
```

```python
df.info()
```

```python
df.date_of_birth = pd.to_datetime(df.date_of_birth,dayfirst=True)
df
```

```python
df.info()
```

```python
today = pd.to_datetime("04-05-2022",dayfirst=True)
today
```

```python
def get_age(dob):
    days = (today-dob).days
    age = days/365
    age = np.floor(age)
    return int(age)
```

```python
df['age'] = df.date_of_birth.apply(get_age)
```
