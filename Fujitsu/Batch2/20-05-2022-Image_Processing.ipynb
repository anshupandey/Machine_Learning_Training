## Images
    - images are composed of pixels, images can be thought of as collection of pixels
    - each pixel represents a color, one pixel = one color box
    - primarily images are processed in 8 bit size. that means values in pixels can vary from 0000 0000 (0) to 1111 1111 (255)
    - images can be of two types
        - GrayScale
        - Color (RGB)

### Mophological Analysis
    - Erosion - loss values to the pixels
    - Dilation - gain of values to the pixels

```python
img = plt.imread(r"E:\image_data\np2.jpg")
img.shape
```

```python
plt.imshow(img)
plt.show()
```

    - higher resolutions = high kernel size / (here kernel is referred as variable factor)
    - lower resolutions = low kernel size 

```python
# erosion
factor = np.ones((5,5))
img2 = cv2.erode(img,factor)
plt.imshow(img2)
plt.show()
```

```python
# erosion
factor = np.ones((7,7))
img2 = cv2.erode(img,factor)
plt.imshow(img2)
plt.show()
```

```python
# dilation
factor = np.ones((7,7))
img3 = cv2.dilate(img2,factor)
plt.imshow(img3)
plt.show()
```

## Color detection

```python
img =plt.imread(r"E:\image_data\dhonikohli2.jpg")
plt.imshow(img)
plt.show()
```

```python
def color_detect(img,rt,gt,bt):
    img2 = img.copy()
    row,col,ch = img.shape
    
    for i in range(0,row):
        for j in range(0,col):
            r,g,b=img[i,j,:]
            if rt-30<r<rt+30 and gt-30<g<gt+30 and bt-30<b<bt+30:
                img2[i,j,:]=[255,255,255]
    return img2
```

```python
imm = color_detect(img,20,20,20)
plt.imshow(imm)
plt.show()
```

```python
imm = color_detect(img,200,150,60)
plt.imshow(imm)
plt.show()
```

```python
%matplotlib qt
plt.imshow(img)
plt.show()
```

```python
%matplotlib inline
plt.imshow(img)
plt.show()
```

```python
imm = color_detect(img,220,130,50)
plt.imshow(imm)
plt.show()
```

## Using haarcascade

```python
fd = cv2.CascadeClassifier(r"E:\image_data\haarcascade_frontalface_alt.xml")
```

```python
img =plt.imread(r"E:\image_data\dhonikohli2.jpg")
plt.imshow(img)
plt.show()
```

```python
corners = fd.detectMultiScale(img)
corners
```

```python
for (x,y,w,h) in corners:
    cv2.rectangle(img,(x,y),(x+w,y+h),[255,0,0],5)
plt.imshow(img)
plt.show()
```
