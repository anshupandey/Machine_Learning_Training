# Hypothesis Testing

A drug production company introduces a new drug ABC and based on testing/research they have done, they claim that this drug can cure a specific disease in 14 days. The business also claims that it cures 95% of patients in 14 days.

We need to verify their claim, we test this drug on 20 patients, below is the num of days it took for them to recover
Days = [12,14,12,15,18,12,10,11,13,15,14,12,15,14,17,15,16,15,16,14]

Objective: compare a sample mean with a population mean (a hypothetical value) 

Z test - it is used to compare a population mean with a sample mean = requires pop std to be known
Null Hypothesis = sample mean is similar to population mean

Popmean = 14
CI = 95%, alpha = 0.05

Population standard deviation (sigma) = not known

T test - it is used to compare a population with a sample mean - if population standard deviation is not known




    Null Hypothesis = sample mean is similar to population mean
    
    - if pvalue > alpha = we accept the null hypothesis - sample mean is similar to popmean
    - if pvalue < alpha = we reject null hypothesis - sample mean is different than popmean

# trial 1

```python
days = [12,14,12,15,18,12,10,11,13,15,14,12,15,14,17,15,16,15,16,14]
popmean = 14
```

```python
from scipy import stats
tscore,pvalue = stats.ttest_1samp(days,popmean)
print(pvalue)
```

```python
tscore
```
