# Image Processing fundamentals

    - image are composed of pixels,
    - primarily there are two types of images - grayscale, color (RGB)
    - Grayscale images
        - 1D images, one pixel has one value
        - 8 bit in size - 00000000 to 11111111
        - in decimal value of each pixel changes from 0 to 255
        - where 0 = black color pixel, 255 = white color pixel

## Mophological Analysis

- Dilation - gain of values to the pixels
- Erosion - loss of values to the pixels

```python
img = plt.imread(r"E:\image_data\np2.jpg")
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
# Erosion
factor = np.ones((7,7))
img2 = cv2.erode(img,factor)
print(img2.shape)
plt.imshow(img2)
plt.show()
```

```python
# Dilation
factor = np.ones((7,7))
img3 = cv2.dilate(img2,factor)
print(img3.shape)
plt.imshow(img3)
plt.show()
```

### Color Detection

```python
def color_detect(img,rt,gt,bt):
    img2 = img.copy()
    row,col,ch = img.shape
    for i in range(0,row,2):
        for j in range(0,col):
            r,g,b = img[i,j,:]
            if (rt-50)<r<(rt+50) and (gt-50)<g<(gt+50) and (bt-50)<b<(bt+50):
                img2[i,j,:] = [255,255,255]
    return img2
```

```python
img = plt.imread(r"E:\image_data\dhonikohli2.jpg")
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
img2 = color_detect(img,30,30,30)
plt.imshow(img2)
plt.show()
```

```python
img2 = color_detect(img,180,50,50)
plt.imshow(img2)
plt.show()
```

## Face detection with openCV

```python
fd = cv2.CascadeClassifier(r"E:\image_data\haarcascade_frontalface_alt.xml")
img = plt.imread(r"E:\image_data\happy3.jpg")
img.shape
```

```python
corners = fd.detectMultiScale(img)
corners
```

```python
for (x,y,w,h) in corners:
    cv2.rectangle(img,(x,y),(x+h,y+w),[255,0,0],2)
plt.imshow(img)
plt.show()
```

```python
img = plt.imread(r"E:\image_data\dhonikohli2.jpg")
corners = fd.detectMultiScale(img)
for (x,y,w,h) in corners:
    cv2.rectangle(img,(x,y),(x+h,y+w),[255,0,0],3)
plt.imshow(img)
plt.show()
```
