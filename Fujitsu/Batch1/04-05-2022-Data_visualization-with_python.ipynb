# Data Visualization

- matplotlib - used for basic data visualization
- seaborn - built on the top of matplotlib and it is used for advance data visualization
- plotly - interactive data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temperature = [25,28,29,34,38,39,44,46,48,44,46,48,43,38,36,35,34,28,29,22,24]
plt.plot(temperature)
plt.show()
```

```python
len(temperature)
```

```python
humidity = [10,15,14,18,19,12,10,5,8,9,8,5,8,5,8,10,12,15,12,12,18]
len(humidity)
```

```python
plt.plot(temperature)
plt.plot(humidity)
plt.show()
```

```python
plt.plot(temperature)
plt.show()
plt.plot(humidity)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(temperature)
plt.plot(humidity)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(temperature,humidity)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(temperature,c='red',label='temperature')
plt.plot(humidity,c='orange',label='humidity')
plt.title("Temperaure and humidity")
plt.xlabel("day number")
plt.ylabel("temperature and humidity")
plt.xlim([0,25])
plt.ylim([0,60])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(temperature,humidity)
plt.show()
```

```python
import numpy as np
x = np.arange(len(temperature))
x
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x,temperature,c='r',label='temperature_discrete')
plt.scatter(x,humidity,c='g',label='humidity_discrete')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(temperature,c='lightblue',label='temperature')
plt.plot(humidity,c='orange',label='humidity')
plt.scatter(x,temperature,c='r',label='temperature_discrete')
plt.scatter(x,humidity,c='g',label='humidity_discrete')
plt.title("Temperaure and humidity")
plt.xlabel("day number")
plt.ylabel("temperature and humidity")
plt.xlim([0,25])
plt.ylim([0,60])
plt.legend()
plt.show()
```

```python
cities = ['DEL','BOM','KUL','DXB','CCU','MAA','BLR']
data = [25,27,28,38,45,56,42]

plt.figure(figsize=(12,4))
plt.bar(cities,data)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.bar(cities,data,color=['orange','red','lightgreen','lightblue','yellow'])
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(data,labels=cities,autopct=str.format("%.2f"),pctdistance=1.3,
       explode=[0.2,0,0,0,0,0,0])
plt.show()
```
