# Image Processing fundamentals

    - image are composed of pixels,
    - primarily there are two types of images - grayscale, color (RGB)
    - Grayscale images
        - 1D images, one pixel has one value
        - 8 bit in size - 00000000 to 11111111
        - in decimal value of each pixel changes from 0 to 255
        - where 0 = black color pixel, 255 = white color pixel
        
            
      
           

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0,255,0,255],
             [0,255,0,255],
             [0,255,0,255],
             [0,255,0,255]])
x.shape
```

```python
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,255,0,255],
             [255,0,255,0],
             [0,255,0,255],
             [255,0,255,0]])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[255,255,255,255],
             [255,0,0,255],
             [255,255,255,255],
             [255,0,0,255]])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,50,100,125],
             [225,200,175,150],
             [250,255,0,255],
             [255,0,255,0]])
plt.imshow(x,cmap='gray')
plt.show()
```


    - Color Images
           - 3D images, one pixel has 3 values
           - each value represents Red, Green and Blue
           - [0,0,0] = black color
           - [255,0,0] = Red color
           - [0,255,0] = Green Color
           - [255,255,255] = white color
           - [0,0,255] = blue color
           - [255,255,0] = yellow color (*)

```python
xx = np.array([[[255,0,0],[0,0,255]],
               [[0,255,0],[0,0,0]],
               [[255,255,0],[255,255,255]],
               [[250,180,0],[0,255,255]],
              ])
```

```python
xx.shape
```

```python
plt.imshow(xx)
plt.show()
```

```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
# load an image
img = cv2.imread(r"E:\image_data\dhoni126.jpg")
img.shape
```

```python
cv2.imshow("MyImage",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
plt.show()
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```

```python
img3 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img3.shape)
plt.imshow(img3,cmap='gray')
plt.show()
```
