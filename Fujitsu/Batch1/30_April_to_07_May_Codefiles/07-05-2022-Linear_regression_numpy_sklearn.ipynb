# Linear Regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,1.8,2.4,3.6,5.4,7.5,8.4,9.5,4.2,6.2,6.8,2.8,2.9,2.4,1.8,2.9,3.4,4.6,5.9,6.4]
price= [180,210,220,320,490,850,890,820,360,520,630,340,380,360,120,240,390,400,480,580]
len(area),len(price)
```

```python
# linear regression
x = np.array(area)
y = np.array(price)

m = 50
c = 20
LR = 0.0015

for i in range(400):
    # get predictions
    yhat = m*x + c
    
    # get error
    error = np.mean(np.square(y-yhat))
    
    # apply gradient descent
    # get the partial derivatives (gradient)
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    # Apply gradient descent to update m and c
    m = m - LR*dedm
    c = c - LR*dedc
    print(error)
```

```python
print(m,c)
```

# Linear Regression with scikit-learn (sklearn)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,1.8,2.4,3.6,5.4,7.5,8.4,9.5,4.2,6.2,6.8,2.8,2.9,2.4,1.8,2.9,3.4,4.6,5.9,6.4]
price= [180,210,220,320,490,850,890,820,360,520,630,340,380,360,120,240,390,400,480,580]
len(area),len(price)
```

```python
x = np.array(area).reshape(20,1)
y = np.array(price)
```

### OLS implementation 

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# train the model with training data
model.fit(x,y)
```

```python
print("m and c ",model.coef_,model.intercept_)
```

```python
yhat = model.predict(x)
plt.scatter(x,y)
plt.plot(x,yhat,c='r')
plt.show()
```

### Gradient Descent Implementation

```python
from sklearn.linear_model import SGDRegressor
model2 = SGDRegressor(learning_rate='invscaling',
                     tol=0.0001,max_iter=2000,random_state=3)
model2.fit(x,y)
```

```python
print("m and c ",model2.coef_,model2.intercept_)
```

```python
yhat = model2.predict(x)
plt.scatter(x,y)
plt.plot(x,yhat,c='r')
plt.show()
```
