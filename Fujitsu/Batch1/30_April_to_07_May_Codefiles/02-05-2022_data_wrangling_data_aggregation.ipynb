# Data Wrangling, transformation

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\AI\data\datasets-1\regiment.csv")
df1.head()
```

```python
df1 = pd.read_csv(r"D:\AI\data\datasets-1\regiment.csv",index_col='index')
df1.head()
```

```python
df2 = pd.read_csv(r"D:\AI\data\datasets-1\regiment2.csv")
df2.head()
```

```python
df1
```

```python
df = pd.merge(left=df1,right=df2,on='name',how='inner')
df
```

```python
df.head()
```

```python
# creating a new column
df['improvement'] = df['postTestScore'] - df['preTestScore']
df.head()
```

```python
# Binnig - converting a column from numeric to categoric
df['Impr'] = pd.cut(df.improvement,
                    bins=[0,30,60,120],
                    labels=['low','medium','high'])
df.head()
```

```python
# binning with qcut
df['impr2'] = pd.qcut(df.improvement,q=4,labels=['A','B','C','D'])
df
```

```python
df['imp_pct'] = df.improvement.rank(pct=True)
df
```

```python
df.improvement.max() - df.improvement.min()
```

```python
# dropa column
df.drop(columns=['impr2','Impr'],inplace=True)
df
```

## Apply method

```python
df
```

```python
def convert(a):
    return a.upper()
```

```python
df['name'] = df['name'].apply(convert)
df
```

```python
df.info()
```

```python
# date of birth to age
# first step is to standardize the date format
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
df.info()
```

```python
import numpy as np
```

```python
today = pd.to_datetime("30-04-2022")

def getage(dob):
    days = (today - dob).days
    age = days/365
    age = np.floor(age)
    return age
```

```python
df['age'] = df['date_of_birth'].apply(getage)
df
```
