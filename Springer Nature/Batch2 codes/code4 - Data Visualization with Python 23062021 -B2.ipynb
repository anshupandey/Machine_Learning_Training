# Data Visualization with Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10,0.5)
y = 2*x + x**2
z = 5*x + x**1.5
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,c='r')
plt.plot(x,z,c='g')
plt.show()
```

```python
plt.plot(x,y,c='r')
plt.show()
plt.plot(x,z,c='g')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,c='r',label='x vs y')
plt.plot(x,z,c='g',label='x vs y')
plt.xlabel("values of x")
plt.ylabel("values of y")
plt.title("Graph between x,y and z")
plt.legend()
plt.show()
```

### Scatter plot

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y)
plt.plot(x,y,c='r')
plt.show()
```

### Bar plot

```python
names = ['CCU','MAA','BLR','BOM','PNQ','DEL']
vals = [45,56,23,45,42,32]
plt.figure(figsize=(12,5))
plt.bar(names,vals,color=['red','lightgreen','orange','lightblue'])
plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(12,5))
plt.pie(vals,labels=names)
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(12,5))
plt.boxplot(x)
plt.show()
```
