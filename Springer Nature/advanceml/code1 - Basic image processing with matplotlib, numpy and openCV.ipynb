### Images
    - images consists of matrix of pixels
    - pixels are arranged in rows and columns
    - usually all tools process images in 8 bit size
    
    - Grayscale images  - one pixel is represented using a value of 8 bits
        - lowest value - 0000 0000 - 0 - black
        - highest value = 1111 1111 - 255 - white
    - Color images - RGB images - one pixels is represented using 3 values of 8 bits each - (red,green,blue)

```python
x = np.array([[255,0,255,0],
             [255,0,255,0],
             [255,0,255,0],
             [255,0,255,0]])
print(x.shape)
```

```python
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[255,0,255,0],
             [0,255,0,255],
             [255,0,255,0],
             [0,255,0,255]])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,30,62,90],
             [110,135,160,185],
             [205,220,245,255],
             [0,255,0,255]])
plt.imshow(x,cmap='gray')
plt.show()
```

## Morphological analysis

    - Erosion - loss of values to the pixels -> pixels with higher values become lower
    - Dilation - gain of pixel values -> pixels with lower values become higher

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = plt.imread(r"E:\image_data\np2.jpg")
plt.imshow(img)
plt.show()
```

```python
# Erosion
kernel = np.ones((7,7))
img2 = cv2.erode(img,kernel)
plt.imshow(img2)
plt.show()
```

```python
# Dilation
img3 = cv2.dilate(img2,kernel)
plt.imshow(img3)
plt.show()
```

## Working with haarcascade models

```python
img = plt.imread(r"E:\image_data\dhonikohli2.jpg")
plt.imshow(img)
plt.show()
```

```python
fd = cv2.CascadeClassifier(r"E:\image_data\haarcascade_frontalface_alt.xml")
```

```python
img.shape
```

```python
corners = fd.detectMultiScale(img)
corners
```

```python
for (x,y,w,h) in corners:
    cv2.rectangle(img,(x,y),(x+w,y+h),[255,0,0],2)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
plt.imshow(img2)
plt.subplot(223)
plt.imshow(img3)
plt.subplot(224)
plt.imshow(img)
plt.show()
```
