# Image processing with numpy, matplotlib and openCV

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Images
    - images consists of matrix of pixels
    - pixels are arranged in rows and columns
    - usually all tools process images in 8 bit size
    
    - Grayscale images  - one pixel is represented using a value of 8 bits
        - lowest value - 0000 0000 - 0 - black
        - highest value = 1111 1111 - 255 - white
    - Color images - RGB images - one pixels is represented using 3 values of 8 bits each - (red,green,blue)

```python
x = np.array([[255,0,255,0],
             [255,0,255,0],
             [255,0,255,0],
             [255,0,255,0]])
print(x.shape)
```

```python
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[255,0,255,0],
             [0,255,0,255],
             [255,0,255,0],
             [0,255,0,255]])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,30,62,90],
             [110,135,160,185],
             [205,220,245,255],
             [0,255,0,255]])
plt.imshow(x,cmap='gray')
plt.show()
```

## Color Images - R,G,B
    - [255,0,0] = Red
    - [0,255,0] = Green
    - [0,0,255] - Blue
    - [0,0,0] - Black
    - [255,255,255] - White
    - [255,255,0] = yellow

```python
y = np.array([[[0,0,255],[255,0,0],[255,255,0]],
              [[0,255,0],[0,0,0],[0,255,100]]])

plt.imshow(y)
plt.show() 
```

## Controlling brightness

```python
kernel = np.ones(img.shape,dtype='int32')
img2 = img + 50*kernel

plt.imshow(img2)
plt.show()
```

```python
img2 = img - 50*kernel

plt.imshow(img2)
plt.show()
```
