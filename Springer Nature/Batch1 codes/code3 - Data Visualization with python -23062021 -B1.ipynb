# Data VIsualization with Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10,0.5)
y = np.exp(x)
z = np.exp(x-0.2)
```

```python
x
```

### Line plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,c='r')
plt.plot(x,z,c='g')
plt.show()
```

```python
plt.plot(x,y,c='r')
plt.show()
plt.plot(x,z,c='g')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,c='r',label='x vs y')
plt.plot(x,z,c='g',label='x vs z')
plt.xlabel("Values of x")
plt.ylabel("values of y and z")
plt.title("graph of x,y,z")
plt.legend()
plt.show()
```

### Scatterplot

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y)
plt.plot(x,y,c='r')
plt.show()
```

### barplot

```python
names = ['CCU','BOM','MAA','BLR','DEL']
vals = [45,56,12,23,78]
plt.figure(figsize=(12,5))
#plt.bar(names,vals,color=['red','lightgreen','lightblue','orange','yellow'])
plt.bar(names,vals,color=['r','g','y','b'])
plt.show()
```

### Pie chart

```python
plt.pie(vals,labels=names)
plt.show()
```

### Boxplot

```python
plt.boxplot(x)
plt.show()
```
