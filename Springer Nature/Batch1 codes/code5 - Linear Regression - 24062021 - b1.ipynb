# Linear Regression with numpy

<img src="linear_regression.png">

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,2.4,3.6,2.5,1.8,3.6,4.8,5.6,9.2,8.3,8.4,7.6,6.7,5.4,4.9]
price= [150,210,310,290,320,395,510,630,870,820,790,710,620,590,540]
```

```python
x = np.array(area)
y = np.array(price)
```

```python
len(x),len(y)
```

```python
LR = 0.002
m = 10
c = 20

for i in range(1000):
    yhat = m*x + c
    
    # get error
    e = np.mean(np.square(y-yhat))
    
    # apply gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(e)
    
    #plt.plot(x,yhat,c='g')
    #plt.scatter(x,y,c='r')
    #plt.show()
```

```python
print(m,c)
```

# Linear Regression with sklearn

```python
x = np.array(area).reshape(15,1)
y = np.array(price)
```

```python
from sklearn.linear_model import LinearRegression,SGDRegressor
model = LinearRegression() # creating an object model of type class linear regression
model.fit(x,y)
```

```python
print(model.coef_,model.intercept_)
```

```python
# make predictions
model.predict([[4.3]])
```

## ###################################m

```python
model = SGDRegressor(learning_rate='adaptive',max_iter=4000,tol=1e-10,random_state=5)
model.fit(x,y)
```

```python
print(model.coef_,model.intercept_)
```
