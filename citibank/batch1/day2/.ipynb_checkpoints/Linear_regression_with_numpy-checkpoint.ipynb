# Linear Regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
area = [1.2,2.4,5.6,8.7,5.9,6.8,5.4,2.9,3.6,5.8,1.8,2.5,8.4,5.6,6.2]
price =[190,280,480,760,680,720,570,220,420,490,200,200,890,680,720]
len(area),len(price)
```

```python
x = np.array(area)
y = np.array(price)
m = 10
c = 20
LR = 0.002

for i in range(5):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    e = np.mean(np.square(y-yhat))
    
    # gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(e)
```

```python
x = np.array(area)
y = np.array(price)
m = 10
c = 20
LR = 0.002

for i in range(10):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    e = np.mean(np.square(y-yhat))
    
    # gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(e)
    
    plt.scatter(x,y,c='g')
    plt.plot(x,yhat,c='r')
    plt.show()
```

```python
x = np.array(area)
y = np.array(price)
m = 10
c = 20
LR = 0.002

for i in range(1000):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    e = np.mean(np.square(y-yhat))
    
    # gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(e)
```

```python
print(m,c)
```

```python
plt.scatter(x,y)
plt.plot(x,yhat,c='r')
plt.show()
```

# Linear regression with scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
area = [1.2,2.4,5.6,8.7,5.9,6.8,5.4,2.9,3.6,5.8,1.8,2.5,8.4,5.6,6.2]
price =[190,280,480,760,680,720,570,220,420,490,200,200,890,680,720]
len(area),len(price)
```

```python
x = np.array(area).reshape(15,1)
y = np.array(price)
```

```python
from sklearn.linear_model import SGDRegressor
model = SGDRegressor(learning_rate='invscaling',tol=0.000001)
model.fit(x,y)
```

```python
print("m and c are ",model.coef_,model.intercept_)
```

## OLS implementation

```python
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
model2.fit(x,y)
print(model2.coef_,model2.intercept_)
```

```python
yhat = model.predict(x)
plt.scatter(x,y)
plt.plot(x,yhat,c='r')
plt.show()
```
