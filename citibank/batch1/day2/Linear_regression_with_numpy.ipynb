# Linear Regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
area = [1.2,2.4,5.6,8.7,5.9,6.8,5.4,2.9,3.6,5.8,1.8,2.5,8.4,5.6,6.2]
price =[190,280,480,760,680,720,570,220,420,490,200,200,890,680,720]
len(area),len(price)
```

```python
x = np.array(area)
y = np.array(price)
m = 10
c = 20
LR = 0.002

for i in range(5):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    e = np.mean(np.square(y-yhat))
    
    # gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(e)
```
