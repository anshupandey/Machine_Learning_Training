# Image Processing Fundamentals

```python
import numpy as np
import matplotlib.pyplot as plt
```

    Images consists of pixels - images can be thought of as matrices of pixels
    Usually images are processed in 8 bit size.
    
    Prmarily images can be of two types - 
    - Grayscale images - one pixel is represented with one value 
        - the vlaue of pixel can range from 0000 0000 to 1111 1111 (0 to 255)
        
    - RGB (color) images - one pixel is represented with 3 values - Red, Green and Blue

```python
x = np.array([[0,255,0,255],
              [0,255,0,255],
             [0,255,0,255],
             [0,255,0,255]])
print(x.shape)
```

### Color images
    - each pixels consists of 3 values - R,G,B
    
    - [0,0,0] = Black
    - [255,0,0] = Red
    - [0,255,0] = Green
    - [0,0,255] = Blue
    - [255,255,0] = Yellow
    - [255,255,255] = White

```python
y = np.array([[[0,0,0],[255,0,0]],
             [[0,255,0],[255,100,0]],
             [[0,0,255],[255,0,255]],
             [[255,255,0],[255,255,255]]])
y.shape
```

```python
# standard shape of image = rows, cols, channel
# channel = depth of image / number of values with in 1 pixel
# for grayscale image, channel = 1
# for color RGB images, channel = 3
```

```python
plt.imshow(y)
plt.show()
```

## Morphological Analysis
    - Erosion - loss of values to the pixels on the boundary ( the pixels at the boundary of the object loose values)
        - the pixel value is replaced by lowest value in the neughbourhood (neighbourhood is decided based on structuring element)
        
    - Dilation - gain of values to the pixels on the boundary ( the pixels at the boundary of the object gain values)
        -- the pixel value is replaced by highest value in the neughbourhood (neighbourhood is decided based on structuring element)

```python
img = plt.imread(r"E:\image_data\np2.jpg")
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
kernel = np.ones((7,7),dtype='int32')
# Erosion
img2 = cv2.erode(img,kernel)
plt.imshow(img2)
plt.show()
```

```python
kernel
```

```python
# Dilation
img3 = cv2.dilate(img2,kernel)
plt.imshow(img3)
plt.show()
```
