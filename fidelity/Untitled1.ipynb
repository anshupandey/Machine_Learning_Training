```python
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## OCR - Optical Character Recognition

```python
img = plt.imread(r"D:\AI\fidelity\img.jpg")
img.shape
```

```python
plt.figure(figsize=(40,10))
plt.imshow(img)
plt.show()
```

```python
text = pytesseract.image_to_string(img)
print(text)
```

```python
grayimg = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img2 = cv2.threshold(grayimg,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
img4 = cv2.threshold(grayimg,200,255,cv2.THRESH_BINARY)[1]
```

```python
plt.figure(figsize=(40,10))
plt.imshow(img4,cmap='gray')
plt.show()
```

```python
text = pytesseract.image_to_string(img4)
print(text)
```

```python
h,w,c = img.shape
boxes = pytesseract.image_to_boxes(img)
print(boxes)
```

```python
for obj in boxes.splitlines():
    obj = obj.split(" ")
    img = cv2.rectangle(img,(int(obj[1]),h-int(obj[2])),(int(obj[3]),h-int(obj[4])),(255,0,0),2)
plt.figure(figsize=(40,10))
plt.imshow(img)
plt.show()
```
