# Time series forecasting

## Passenger volume forecasting for singapore changi airport

1. Standardize the time series data

    - handle noisy data
    - Normalize the time series
    

2. Time Series EDA

3. Apply forecasting technique

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# load data
df = pd.read_csv(r"D:\AI\collabera\thoughtspot\changi_airport_data.csv")
df.shape
```

```python
df.head()
```

```python
df = df[df.level_1=="Total Passengers"]
df.index = pd.to_datetime(df.month)
df = df[['value']]
df.head()
```

```python
# data from 2004 to 2018
df2 = df['2004-01-01':'2018-12-01']
df2.shape
```

```python
plt.figure(figsize=(12,5))
plt.plot(df2)
plt.show()
```

```python
# Normalize time series
dflog = np.log(df2)
plt.figure(figsize=(12,5))
plt.plot(dflog)
plt.show()
```

```python
# Decomposing time series components
from statsmodels.tsa.seasonal import seasonal_decompose


decompose = seasonal_decompose(df2,model='additive',period=12)
```

```python
fig = decompose.plot()
fig.set_size_inches((10,8))
fig.tight_layout()
plt.show()
```

#### Check for stationarity: Dicky fuller test
 
    Stationarity = no trend
    
    Null Hypothesis: Data is not stationary, it has trend
    
    Considering confidence interval = 95%, alpha = 1 - 0.95 = 0.05
    
    - if pvalue > alpha = we accept null hypothesis, data is not stationary and it has trend
    - if pvalue < alpha = we reject null hypothesis, data is stationary, it has no trend

```python
from statsmodels.tsa.stattools import adfuller
adfuller(df2['value'])
```

```python
dflog.head()
```

```python
dflog.shift(1).head()
```

```python
dflogdiff = dflog - dflog.shift(1)
plt.figure(figsize=(12,5))
plt.plot(dflogdiff)
plt.show()
```

```python
adfuller(dflogdiff['value'].dropna())
```

## Time series forecasting with ARIMA

```python
import pmdarima
```

```python
from pmdarima.arima import auto_arima
```

```python
model = auto_arima(dflog,suppress_warnings=True,seasonal=False,trace=True,
                  start_p=1,start_q=1,d=1,max_p=3,max_q=3)
model.fit(dflog)
```

```python
test = df['2019-01-01':'2019-12-01']
test
```

```python
forecast = model.predict(n_periods=12)
forecast = pd.DataFrame(np.exp(forecast,),index = test.index)
forecast
```

```python
plt.figure(figsize=(12,5))
plt.plot(test,c='green')
plt.plot(forecast,c='red')
plt.show()
```

```python
# MAPE - mean absolute percentage error
from sklearn import metrics
metrics.mean_absolute_percentage_error(test,forecast)
```

## Time series forecasting with SARIMA: Seasonal ARIMA

    p,d,q = ARIMA params
    P,D,Q = seasonal params
    n = cycle size

```python
model = auto_arima(dflog,suppress_warnings=True, seasonal=True, 
                  start_p=1,d=1,start_q=1,max_p=3,max_q=3,
                  start_P=0,start_Q=0,max_P=3,max_Q=3,m=12,
                  trace=True)

model.fit(dflog)
```

```python
test = df['2019-01-01':'2019-12-01']
test
```

```python
forecast = model.predict(n_periods=12)
forecast = pd.DataFrame(np.exp(forecast,),index = test.index)
forecast
```

```python
plt.figure(figsize=(12,5))
plt.plot(test,c='green')
plt.plot(forecast,c='red')
plt.legend(['Actual','Forecast'])
plt.show()
```

```python
# MAPE - mean absolute percentage error
from sklearn import metrics
metrics.mean_absolute_percentage_error(test,forecast)
```
