## Linear Regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,1.8,2.4,3.6,5.4,7.5,8.4,9.5,4.2,6.2,6.8,2.8,2.9,2.4,1.8,2.9,3.4,4.6,5.9,6.4]
price= [180,210,220,320,490,850,890,820,360,520,630,340,380,360,120,240,390,400,480,580]
len(area),len(price)
```

```python
x = np.array(area)
y = np.array(price)
m = 40
c = 30
LR = 0.0015

for i in range(5):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y-yhat))
    
    # Apply Gradient descent to minimize the error
    
    # calculate gradients
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(error)
    plt.scatter(x,y)
    plt.plot(x,yhat,c='red')
    plt.show()
```

```python
y = np.array(price)
m = 40
c = 30
LR = 0.0015

for i in range(500):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y-yhat))
    
    # Apply Gradient descent to minimize the error
    
    # calculate gradients
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(error)
```

```python
print(m,c)
```

## Linear Regression with scikit-learn

```python
area = [1.2,1.8,2.4,3.6,5.4,7.5,8.4,9.5,4.2,6.2,6.8,2.8,2.9,2.4,1.8,2.9,3.4,4.6,5.9,6.4]
price= [180,210,220,320,490,850,890,820,360,520,630,340,380,360,120,240,390,400,480,580]
len(area),len(price)
```

```python
x = np.array(area).reshape(20,1)
y = np.array(price)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
# train the model
model.fit(x,y)
```

```python
print("m and c values are ",model.coef_,model.intercept_)
```

```python
yhat = model.predict(x)

plt.scatter(x,y)
plt.plot(x,yhat,c='red')
plt.show()
```

## Sklearn linear regression gradient descent variant

```python
from sklearn.linear_model import SGDRegressor
model = SGDRegressor(learning_rate='invscaling',tol=0.0000001, max_iter=2000,penalty='l1',l1_ratio=0.8)
# train the model
model.fit(x,y)
```

```python
print("m and c values are ",model.coef_,model.intercept_)
```

```python
yhat = model.predict(x)

plt.scatter(x,y)
plt.plot(x,yhat,c='red')
plt.show()
```
